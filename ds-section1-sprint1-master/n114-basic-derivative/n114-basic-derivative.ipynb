ì—¬ê¸°ì„œ ì´ ê³¼ì •ì€ $f(a,b) = \varepsilon$ ë¡œ í‘œí˜„ ë  ìˆ˜ ìˆìœ¼ë©°, ì˜¤ì°¨ í•¨ìˆ˜ì¸ $\varepsilon$ì„ ìµœì†Œí™” í•˜ëŠ” $a,b$ë¥¼ ì°¾ëŠ” ê²ƒì´ ë¨¸ì‹ ëŸ¬ë‹(`Linear regression`)ì˜ ëª©í‘œì…ë‹ˆë‹¤.

## ë¯¸ë¶„ ê³µì‹ w/ Python

ì„¸ìƒì—ëŠ” ë‹¤ì–‘í•œ í•¨ìˆ˜ê°€ ì¡´ì¬í•˜ê¸° ë•Œë¬¸ì— ë¯¸ë¶„ì„ í•˜ëŠ” ë°©ë²• ë˜í•œ ë§¤ìš° ë§ì´ ì¡´ì¬í•©ë‹ˆë‹¤.

ë¯¸ë¶„ì˜ ì›ë¦¬ì— ëŒ€í•´ ê¹Šê²Œ íŒŒê³ ë“¤ë©´ì„œ ì–´ë–»ê²Œ í•¨ìˆ˜ë¥¼ ë¯¸ë¶„í•´ì•¼í•˜ëŠ”ì§€ ì´í•´í•´ê°€ë©´ì„œ ë„í•¨ìˆ˜ë¥¼ ë„ì¶œí•´ë‚´ëŠ” ê²ƒì€ ì´í›„ì— ì¡°ê¸ˆ ë” ë³µì¡í•œ í•¨ìˆ˜ë¥¼ ë‹¤ë£° ë•Œë¥¼ ëŒ€ë¹„í•´ì„œ ë§¤ìš° ì¢‹ì€ ìƒê° ì…ë‹ˆë‹¤.

í•˜ì§€ë§Œ ì•„ì‰½ê²Œë„ ìš°ë¦¬ì—ê²Œ ì£¼ì–´ì§„ ì‹œê°„ì´ ë„ˆë¬´ë‚˜ë„ ì§§ê¸° ë•Œë¬¸ì— ì˜¤ëŠ˜ì€ ëŒ€í‘œì ì¸ ì§€ë¦„ê¸¸ ëª‡ê°€ì§€ì™€ ì—¬ëŸ¬ë¶„ì´ ì•ìœ¼ë¡œ í•„ìš”í•˜ê²Œ ë  ì¤‘ìš”í•œ ë°©ë²• ëª‡ê°€ì§€ë§Œ ë‹¤ë£¨ë„ë¡ í•˜ê² ìŠµë‹ˆë‹¤.

ê¸°ë³¸ì ìœ¼ë¡œ ë¯¸ë¶„ì€ ë‹¤ìŒì˜ ê³µì‹ì„ í™œìš©í•©ë‹ˆë‹¤: 

  $f'(x) = {f(x + \Delta x) - f(x) \over \Delta x}$, $\Delta x \rightarrow 0$
    
í•˜ì§€ë§Œ ì‹¤ì œë¡œ 0ìœ¼ë¡œ ë‚˜ëˆŒ ìˆ˜ëŠ” ì—†ê¸° ë•Œë¬¸ì— 0ì— ë§¤ìš° ê·¼ì‚¬í•œ ê°’ì„ ì‚¬ìš©í•˜ê²Œ ë©ë‹ˆë‹¤. ë³´í†µ $1e-5$ ì„ ì‚¬ìš©í•˜ë©°, ì´ëŸ¬í•œ ì ‘ê·¼ ë°©ì‹ì„ `numerical method` ë¼ëŠ” ë°©ë²•ìœ¼ë¡œ í‘œí˜„í•˜ê¸°ë„ í•©ë‹ˆë‹¤.

í•œí¸, `numerical method`ì—ì„œëŠ” ì¡°ê¸ˆ ë” ì •í™•í•œ ì¸¡ì •ì„ ìœ„í•´ ë¶„ìì— $f(x + \Delta x) - f(x - \Delta x) \over 2\Delta x$ ë¥¼ ì‚¬ìš©í•˜ê¸°ë„ í•©ë‹ˆë‹¤.

ê·¸ëŸ¼ ë¨¸ì‹ ëŸ¬ë‹ì—ì„œ ëŒ€í‘œì ìœ¼ë¡œ ì“°ì´ëŠ” ë¯¸ë¶„ ê³µì‹ë“¤ì— ëŒ€í•´ ì•Œì•„ë³´ê² ìŠµë‹ˆë‹¤

1.  $f(x)$ = ìƒìˆ˜ $\rightarrow$ $f'(x)$ = 0

f'(x)ê°€ ìƒìˆ˜ (constant)ì¸ ê²½ìš°ì—ëŠ” xë¥¼ ì•„ë¬´ë¦¬ ëŠ˜ë¦¬ê±°ë‚˜ ì¤„ì—¬ë„ ëŠ˜ ê°™ì€ ìˆ«ìì´ê¸° ë•Œë¬¸ì— ë³€í™”ê°€ ì „í˜€ ì—†ìŠµë‹ˆë‹¤. ê·¸ ë§ì€ ì¦‰ ë³€í™”ìœ¨ì´ 0ì´ê¸° ë•Œë¬¸ì— ë¯¸ë¶„ê³„ìˆ˜ë„ ëŠ˜ 0ì…ë‹ˆë‹¤

```python
# ì˜ˆì‹œ 1 : Numerical Method

# f(x) = 5
def f(x):
    return 5

def numerical_derivative(fx, x):
    delta_x = 1e-5

    # return (fx(x + delta_x) - fx(x)) / delta_x
    return ((fx(x + delta_x) - fx(x - delta_x)) / 2*delta_x)

print(numerical_derivative(f, 1))

# ì˜ˆì‹œ 2 : Scipyì˜ derivative í™œìš©
from scipy.misc import derivative

# ë‘ ë°©ë²•ì˜ ê²°ê³¼ê°’ ë¹„êµ
derivative(f,1, dx=1e-6) == numerical_derivative(f, 1)
```

2. $f(x) = ax^{n}$ $\rightarrow$ $f'(x) = an{x}^{(n-1)}$

`Power Rule`ë¡œë„ ì•Œë ¤ì ¸ìˆëŠ” ë¯¸ë¶„ë²•ì…ë‹ˆë‹¤. x ê¸°ì¤€ìœ¼ë¡œ nìŠ¹ì„ ë¯¸ë¶„í•  ê²½ìš° nì„ ë‚´ë ¤ë³´ë‚´ì„œ ê³±í•´ì¤€ í›„, ì´í›„ì— nìŠ¹ì—ì„œ "í•˜ë‚˜"ë¥¼ ê°€ì ¸ì™”ê¸° ë•Œë¬¸ì— ë¹¼ì£¼ëŠ” ë°©ë²•ì…ë‹ˆë‹¤.

ì˜ˆì‹œë¡œ $f(x) = 3x^4 + 10$ ë¥¼ ë¯¸ë¶„í•´ë³´ê² ìŠµë‹ˆë‹¤.

ë¨¼ì € 4ìŠ¹ì—ì„œ í•˜ë‚˜ë¥¼ ë‚´ë ¤ë³´ë‚´ì„œ ì•ì— ìˆëŠ” 3ê³¼ ê³±í•´ì¤ë‹ˆë‹¤. (10ì€ ìƒìˆ˜ì´ê¸° ë•Œë¬¸ì— ë¯¸ë¶„ì„ í•˜ë©´ 0ì´ ë©ë‹ˆë‹¤)<br>
$f'(x) = (4*3)x^4$

ì´í›„ì—ëŠ” 4ìŠ¹ì—ì„œ 1ì„ ë¹¼ì¤ë‹ˆë‹¤ ("ë¹Œë ¤ì¤¬ê¸° ë•Œë¬¸ì—")<br>
$f'(x) = (4*3)x^{4-1}$

ìµœì¢…ì ìœ¼ë¡œ f(x)ì˜ ë„í•¨ìˆ˜ëŠ” ì´ë ‡ê²Œ ê³„ì‚°ë©ë‹ˆë‹¤.<br>
$f'(x) = 12x^3$

x = 2ì¼ ë•Œ f'(x)ëŠ” ë‹¤ìŒê³¼ ê°™ìŠµë‹ˆë‹¤<br>
$f'(2) =  96$

```python
# ê°™ì€ ê²°ê³¼ë¥¼ numerical methodë¥¼ í™œìš©í•´ì„œ ê³„ì‚°í•´ë³´ê² ìŠµë‹ˆë‹¤
def f(x):
    return 3*(x**4) + 10

def numerical_derivative(fx, x):
    delta_x = 1e-5

    # return (fx(x + delta_x) - fx(x)) / delta_x
    return ((fx(x + delta_x) - fx(x - delta_x)) / 2*delta_x)

print(numerical_derivative(f, 2))

# ì˜ˆì‹œ 2 : Scipyì˜ derivative í™œìš©
from scipy.misc import derivative

# ë‘ ë°©ë²•ì˜ ê²°ê³¼ê°’ ë¹„êµ
print(derivative(f,2, dx=1e-5))
derivative(f,2, dx=1e-5) == numerical_derivative(f, 2)  #delta Xì˜ ê°’ì€ ê°™ì§€ë§Œ Rounding ì—ëŸ¬ë¡œ ì¸í•´ ë‘ ê²°ê³¼ê°€ ë¯¸ë¬˜í•˜ê²Œ ë‹¤ë¥¸ ê²ƒì„ í™•ì¸í•˜ì‹¤ ìˆ˜ ìˆìŠµë‹ˆë‹¤.
# derivativeëŠ” ((fx(x + delta_x) - fx(x - delta_x)) / 2*delta_x) ê°’ìœ¼ë¡œ ì„¤ì • ë˜ì–´ ìˆë‹¤.
```

## ê²½ì‚¬í•˜ê°•ë²•

ê²½ì‚¬í•˜ê°•ë²• (Gradient Descent, ì´í•˜ GD)ëŠ” ìœ„ì—ì„œ ê±°ë¡  ëë˜ ì˜¤ì°¨ í•¨ìˆ˜ì¸  ğœ€ ì„ ìµœì†Œí™” í•˜ëŠ”  ğ‘,ğ‘ ë¥¼ ì°¾ì„ ìˆ˜ ìˆëŠ” ìµœì í™” ì•Œê³ ë¦¬ì¦˜ ì¤‘ì˜ í•˜ë‚˜ì…ë‹ˆë‹¤. ê°•ì˜ì—ì„œ ìµœì ì˜ a,bë¥¼ ì°¾ê¸° ìœ„í•´ì„  ë¯¸ë¶„ê³„ìˆ˜ê°€ 0ì¸ ê³³ì„ ì°¾ìœ¼ë©´ ëœë‹¤ê³  í–ˆìŠµë‹ˆë‹¤. í•˜ì§€ë§Œ í˜„ì‹¤ì ìœ¼ë¡œ ìš°ë¦¬ê°€ ì•ìœ¼ë¡œ ë‹¤ë£¨ê²Œ ë  ë¬¸ì œì—ì„  íŒŒë¼ë¯¸í„°ì˜ ê°¯ìˆ˜ëŠ” ìˆ˜ì—†ì´ ë§ì„ ê²ƒì´ê³  í•˜ë‚˜ì˜ minimum/maximumë§Œì´ ì¡´ì¬í•˜ì§€ ì•ŠëŠ” ìƒí™©ì— ì§ë©´í•˜ê²Œ ë  ê²ƒì…ë‹ˆë‹¤. 

ê²½ì‚¬í•˜ê°•ë²•ì€ ì„ì˜ì˜ a, bë¥¼ ì„ íƒí•œ í›„ (random initialization)ì— ê¸°ìš¸ê¸° (gradient)ë¥¼ ê³„ì‚°í•´ì„œ ê¸°ìš¸ê¸° ê°’ì´ ë‚®ì•„ì§€ëŠ” ë°©í–¥ìœ¼ë¡œ ì§„í–‰í•©ë‹ˆë‹¤. ê¸°ìš¸ê¸°ëŠ” í•­ìƒ ì†ì‹¤ í•¨ìˆ˜ ê°’ì´ ê°€ì¥ í¬ê²Œ ì¦ê°€í•˜ëŠ” ë°©í–¥ìœ¼ë¡œ ì§„í–‰í•©ë‹ˆë‹¤. ê·¸ë ‡ê¸° ë•Œë¬¸ì— ê²½ì‚¬í•˜ê°•ë²• ì•Œê³ ë¦¬ì¦˜ì€ ê¸°ìš¸ê¸°ì˜ ë°˜ëŒ€ ë°©í–¥ìœ¼ë¡œ ì´ë™í•©ë‹ˆë‹¤.

ê²½ì‚¬í•˜ê°•ë²•ì—ì„œ a,bëŠ” ë‹¤ìŒê³¼ ê°™ì´ ê³„ì‚° ë©ë‹ˆë‹¤ :

$a_{n+1} = a_n - \eta âˆ‡ f(a_n)$ <br>
$b_{n+1} = b_n - \eta âˆ‡ f(b_n)$

 ë°˜ë³µì ìœ¼ë¡œ íŒŒë¼ë¯¸í„° a,bë¥¼ ì—…ë°ì´íŠ¸ í•´ê°€ë©´ì„œ ê·¸ë˜ë””ì–¸íŠ¸($âˆ‡ f$)ê°€ 0ì´ ë  ë•Œê¹Œì§€ ì´ë™ì„ í•©ë‹ˆë‹¤. ì´ ë•Œ ì¤‘ìš”í•œê²Œ ë°”ë¡œ í•™ìŠµë¥  (learning rate, $\eta$)ì¸ë°ìš”. í•™ìŠµë¥ ì´ ë„ˆë¬´ ë‚®ê²Œ ë˜ë©´ ì•Œê³ ë¦¬ì¦˜ì´ ìˆ˜ë ´í•˜ê¸° ìœ„í•´ì„œ ë°˜ë³µì„ ë§ì´ í•´ì•¼ë˜ê³  ì´ëŠ” ê²°êµ­ ìˆ˜ë ´ì— ì‹œê°„ì„ ìƒë‹¹íˆ ê±¸ë¦¬ê²Œ í•©ë‹ˆë‹¤. ë°˜ëŒ€ë¡œ í•™ìŠµë¥ ì´ ë„ˆë¬´ í¬ë©´ ì˜¤íˆë ¤ ê·¹ì†Œê°’ì„ ì§€ë‚˜ì³ ë²„ë ¤ì„œ ì•Œê³ ë¦¬ì¦˜ì´ ìˆ˜ë ´ì„ ëª»í•˜ê³  ê³„ì‚°ì„ ê³„ì† ë°˜ë³µí•˜ê²Œ ë  ìˆ˜ë„ ìˆê¸° ë•Œë¬¸ì— í•™ìŠµë¥ ì„ ì •í•  ë•ŒëŠ” ì‹ ì¤‘í•˜ê²Œ ì •í•˜ì…”ì•¼ í•©ë‹ˆë‹¤.


### ì˜ˆì‹œ

```python
def gradient_descent(X, y, lr = 0.05, epoch = 10):
    
    a, b = 0.33, 0.48 # ì„ì˜ ì„ íƒí•œ íŒŒë¼ë¯¸í„° a, b
    N = len(X) # ìƒ˜í”Œ ê°¯ìˆ˜
    
    for _ in range(epoch):            
        f = y - (a*X + b)
    
        # aì™€ bë¥¼ ì—…ë°ì´íŠ¸ í•©ë‹ˆë‹¤
        a -= lr * (-2 * X.dot(f).sum() / N)
        b -= lr * (-2 * f.sum() / N)        
        
    print('a: '+str(a))
    print('b: '+str(b))
```

```python
# y = 3 x + 5
X = np.array([1, 2, 3, 4, 5])
y = np.array([8, 11, 14, 17, 20])
```

```python
gradient_descent(X,y)  # 10ë²ˆë§Œ ëŒë ¸ì„ ë•Œ
```

```python
gradient_descent(X, y, epoch = 100)  # 100ë²ˆ ë°˜ë³µ
```
