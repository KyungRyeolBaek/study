```python
def draw_window( window, yaxisf, color):
    from scipy.interpolate import interp1d
    fscale2idx = interp1d( yaxisf[::-1], arange(yaxisf.size)[::-1])
    x0 = 0.
    for ii, dx in enumerate(window[0]):
        flow = window[1][ii]
        fhigh = window[2][ii]
        ilow = fscale2idx(flow) if flow > yaxisf[-1] else yaxisf.size - 1
        ihigh = fscale2idx(fhigh) if fhigh < yaxisf[0] else 0
        plot([x0, x0+dx], [ilow, ilow], color=color)
        plot([x0, x0+dx], [ihigh, ihigh], color=color)
        plot([x0, x0], [ilow, ihigh], color=color)
        plot([x0+dx, x0+dx], [ilow, ihigh], color=color)
        x0 = x0 + dx
vpeak = cal_freq.getPeaktrace('q', range(2))
fig=cal_freq.showSignal(vCWT,  vfscale, cmap=mpl.cm.jet, aspect='auto', vmax=0.01)
ax = fig.gca()
plot(vpeak[1], color="white")
draw_window(cal_freq.window['q'], vfscale, "grey")
```

* 관심 영역을 확대했다. 주파수 교정 실험의 경우는 V 자 형에서 벗어난 부분은 주파수 곡선을 잘 못 찾은 것이다.
* 수동으로 윈도우를 잘 지정하고 윈도우 영역내에서만 최대값을 구하게 함으로써 주파수 곡선을 잘 구할 수도 있지만,
이것을 교육 시켜서 주파수 곡선을 잘 구하도록 하고자 한다.
* 이 경우는 V자 형을 찾게할 수도 있지만 이것은 일반적인 반사계 데이터를 처리할 때는 전혀 도움을 못주는 특수한 방법이고,
부드럽게 이어지는 곡선을 선호한다는 등 그런 경향을 교육할 수 있으면 반사계 데이터 처리의 간단한 경우로 의의가 있을 것이다.

```python
import tensorflow as tf
from tensorflow.contrib.layers import flatten

tf.reset_default_graph()

def test_cnn(CWT):
    mu = 0
    sigma = 0.1
    # Layer1
    # 10*50*50 -> 5*5 filter, 2channel, 10개 -> stride 1, no padding, bias 10 -> relu
    conv1_w = tf.Variable(tf.truncated_normal(shape = [5,5,2,10], mean = mu, stddev = sigma))  # 5*5 filter, 2 channel, 10개
    conv1_b = tf.Variable(tf.zeros(10))
    conv1 = tf.nn.conv2d(CWT, conv1_w, strides = [1,1,1,1], padding = 'VALID') + conv1_b     # (conv_w, stride 1, no padding) + conv_b
    conv1 = tf.nn.relu(conv1)
    
    # Pooling
    # 10*50*50 -> 5*5 Pooling , ksize [5,5], no padding -> 10*10*10
    # N H W C
    pool_1 = tf.nn.avg_pool(conv1, ksize = [1,5,5,1], strides = [1,5,5,1], padding = 'VALID')  # 5*5 size pooling -> 
    
    # Layer2
    # 10*10*10 -> 2*2 filter, 10, 10 -> stride 1, no padding, bias 8 -> relu -> 10*8*8
    conv2_w = tf.Variable(tf.truncated_normal(shape = [2,2,10,10], mean = mu, stddev = sigma))
    conv2_b = tf.Variable(tf.zeros(10))
    conv2 = tf.nn.conv2d(pool_1, conv2_w, strides = [1,1,1,1], padding = 'VALID') + conv2_b
    conv2 = tf.nn.relu(conv2)
    
    # Pooling2
    # 10*8*8 -> 2*2 Pooling , ksize [2,2] , no padding -> 10*4*4
    pool_2 = tf.nn.avg_pool(conv2, ksize = [1, 2, 2, 1], strides = [1, 2, 2, 1], padding = 'VALID')
    
    # Flatten. 
    # 10*4*4 = 160
    fc1 = flatten(pool_2)

    # Fully Connected layer3
    # 160 -> 50
    fc2_w = tf.Variable(tf.truncated_normal(shape = (160,50), mean = mu, stddev = sigma))
    fc2_b = tf.Variable(tf.zeros(50))
    
## 수정한 곳 !
    fc2 = tf.nn.softmax(tf.matmul(fc1, fc2_w) + fc2_b)
    
#     fc2 = tf.matmul(fc1, fc2_w) + fc2_b
    #Activation
    fc2 = tf.nn.relu(fc2)
    return fc2
```

```python
xvCWT = zeros((1000, 50, 50, 2))
yvCWT = zeros((1000, 50))

# data = 1
d = 1
# for d in range(data):
for n in range(1000):
    num = vpeak[d][25+n]
    yvCWT[n][num] = 1
    for i in range(50):
        for j in range(50):
            realn = vCWT[d][i,j+n].real.astype(float32)
            imgn = vCWT[d][i,j+n].imag.astype(float32)
            xvCWT[n][i][j][0] = sqrt(realn**2 + imgn**2)
            xvCWT[n][i][j][1] = arctan(realn/imgn)
```

```python
# TensorFlow
# Create placeholder
x = tf.placeholder(tf.float32, shape = [None, 50, 50, 2])
y = tf.placeholder(tf.float32, shape = [None, 50])
result = tf.placeholder(tf.float32, shape = [None, 50])
max_result = tf.placeholder(tf.float32, shape = [None])
# Cross_entropy
# yout -> training result , y -> 실제 값
yout = test_cnn(x)
cross_entropy = yout - y

# Loss_Opveration
# cross_entropy
# 넓이를 구해서 줄이기
# loss_operation = tf.reduce_sum(tf.square(cross_entropy))
# softmax
loss_operation = -tf.reduce_sum(y*tf.log(yout))


# Optimizer
# Learing_Rate
rate = 0.001
optimizer = tf.train.AdamOptimizer(learning_rate = rate)

# Training_Operation
training_operation = optimizer.minimize(loss_operation)

init = tf.global_variables_initializer()

sess = tf.Session()
sess.run(init)

# x = [None, 50, 50, 2] , xvCWT = [1000, 50, 50, 2]

for n in range(900):            
    sess.run(training_operation, feed_dict = {x:xvCWT[n:n+100,:,:,:] , y:yvCWT[n:n+100,:]})
    ls = sess.run(loss_operation, feed_dict = {x:xvCWT , y:yvCWT})
    if (n%100 == 0):
        print(ls)
result = sess.run(yout, feed_dict = {x:xvCWT , y:yvCWT})
print(result[0])
```
