```python
f = open("images/sino0.txt", 'r')
line = f.read()
print(line)
f.close
```

```python
import numpy as np
sino_img = np.zeros((256,360))
k = 0
l = 0
for i, j in enumerate(line):
    if j == '\t':
        sino_img[k,l] = float(line[i-14:i])
        k = k+1
        if k == 256:
            k = 0
            l = l+1
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import make_axes_locatable
import numpy as np

plt.imshow(sino_img)
plt.colorbar
```
