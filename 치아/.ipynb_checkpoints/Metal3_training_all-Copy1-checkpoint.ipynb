```python
f = open("images/tooth3-3.txt", 'r')
line = f.read()
print(line)
f.close
```

```python
import numpy as np
tooth3_img = np.zeros((256,360))
k = 0
l = 0
for i, j in enumerate(line):
    if j == '\t':
        tooth3_img[k,l] = float(line[i-14:i])
        k = k+1
        if k == 256:
            k = 0
            l = l+1
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import make_axes_locatable
import numpy as np

plt.imshow(tooth3_img)
plt.colorbar
```

```python
get_ipython().run_line_magic('matplotlib', 'inline')
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import plotly as py
import plotly.graph_objs as go
import requests
from keras.layers import *
from keras.models import *
from keras.callbacks import *
from datetime import datetime
from sklearn.preprocessing import MinMaxScaler
model = Sequential() # Sequeatial Model 
model.add(LSTM(200, input_shape=(1, 5))) # (timestep, feature) 
model.add(Dense(1)) # output = 1 
model.compile(loss='mean_squared_error', optimizer='adam') 
model.summary()
```

```python
result_img = tooth3_img
window_size = 5
x_data = []
y_data = []
k = 0
for j in range(360):
    for i in range(256-window_size-1):
        if 0 in result_img[i:i+1+window_size,j]:
            k=0
        else:
            if result_img[i+1+window_size, j] == 30:
                k=0
            else:
                x_data.append([result_img[i:i+window_size,j]])
                y_data.append(result_img[i+1+window_size,j])

x_train_data = np.asarray(x_data)
y_train_data = np.asarray(y_data)
model.fit(x_train_data, y_train_data, epochs = 10, batch_size = 1)
for j in range(360):
    for i in range(256-window_size-1):
        if 0 in result_img[i:i+1+window_size,j]:
            k=0
        else:
            if result_img[i+1+window_size, j] == 30:
                x_test_data = []
                x_test_data.append([result_img[i:i+window_size,j]])
                x_predict_data = np.asarray(x_test_data)
                prediction = model.predict(x_predict_data)
                result_img[i+1+window_size,j] = prediction[0,0]
```

```python
plt.imshow(result_img)
plt.colorbar
```
