````python
# 설정
import matplotlib.pyplot as plt

plt.rc('font', family='NanumBarunGothic')
plt.rc('axes', unicode_minus=False)
````

````python
import pandas as pd

# url
data_url = "https://ds-lecture-data.s3.ap-northeast-2.amazonaws.com/stocks/Travel.xlsx"

# read_excel
df1= pd.read_excel(data_url, sheet_name=0)
df2= pd.read_excel(data_url, sheet_name=1)
````

````python
# 표 전치 -> 열을 행으로, 행을 열로
df1 = df1.T

# 컬럼 정보 저장 -> 0번째 row에 정보가 있음
df1_columns = df1.iloc[0]

# 컬럼 재 지정
df1.columns=df1_columns

# 필요없는 행(row) 제외
df1 = df1[1:]

# 확인
df1
````

````python
# 표 전치 -> 열을 행으로, 행을 열로
df2 = df2.T

# 컬럼 정보 저장 -> 0번째 row에 정보가 있음
df2_columns = df2.iloc[0]

# 컬럼 재 지정
df2.columns=df2_columns

# 필요없는 행(row) 제외
df2 = df2[1:]

# 확인
df2
````
