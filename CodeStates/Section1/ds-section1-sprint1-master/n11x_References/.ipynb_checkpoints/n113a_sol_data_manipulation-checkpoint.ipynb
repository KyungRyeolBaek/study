````python
# íŒë‹¤ìŠ¤
import pandas as pd
````

````python
# url ì¤€ë¹„
urlhead = 'https://ds-lecture-data.s3.ap-northeast-2.amazonaws.com/stocks/'
url_list = ['000080.csv','000890.csv', '005300.csv', '027740.csv', '035810.csv', '136480.csv']
url_final = []

for i in url_list:
  url = urlhead + i
  url_final.append(url)
````

````python
# urlìœ¼ë¡œ ë°ì´í„°ë¥¼ ë¶ˆëŸ¬ì™€ì„œ ë°ì´í„° í”„ë ˆì„ì— ë‹´ì€ í›„, ì›í•˜ëŠ” ëª¨ì–‘ëŒ€ë¡œ ê°€ê³µí•˜ëŠ” í•¨ìˆ˜ë¥¼ ì •ì˜í•©ë‹ˆë‹¤.
def dataframe_load(url):

  # url ë¶ˆëŸ¬ì˜¤ê¸°
  df_tmp = pd.read_csv(url)
  
  # transpose
  df_tmp = df_tmp.T
  
  # ì²« ë²ˆì§¸ í–‰ì„ ì¸ë±ìŠ¤ë¡œ ì„¤ì •
  new_header = df_tmp.iloc[0]
  df_tmp.columns = new_header

  # ì²« ë²ˆì§¸ í–‰ì„ ì œì™¸í•œ ë‚˜ë¨¸ì§€ í–‰ë“¤ë§Œ ì„ íƒ
  df_tmp = df_tmp[1:]

  # ì™„ì„±ëœ ë°ì´í„° í”„ë ˆì„ì„ ë‚´ê°€ ì§€ì •í•œ ë³€ìˆ˜ì— ì €ì¥í•  ìˆ˜ ìˆë„ë¡ ë°˜í™˜í•©ë‹ˆë‹¤. 
  return df_tmp
````

````python
# ê°ê°ì˜ csvíŒŒì¼ì„ ë°ì´í„° í”„ë ˆì„ìœ¼ë¡œ ë§Œë“¤ì–´ ë³€ìˆ˜ì— ì €ì¥í•©ë‹ˆë‹¤.
df1 = dataframe_load(url_final[0])
df2 = dataframe_load(url_final[1])
df3 = dataframe_load(url_final[2])
df4 = dataframe_load(url_final[3])
df5 = dataframe_load(url_final[4])
df6 = dataframe_load(url_final[5])
````

````python
# í•™ìŠµ ë…¸íŠ¸ì—ì„œ ì“°ë˜ ë°ì´í„°ë¥¼ ê³„ì† ì“°ëŠ” ê²ƒ ì´ë¯€ë¡œ ì¢…ëª©ëª…ì„ ê·¸ëŒ€ë¡œ ê°€ì ¸ì˜µë‹ˆë‹¤.
# csv ë²ˆí˜¸ë¥¼ ì‚´í´ë³´ë‹ˆ 136490ë²ˆì— í•´ë‹¹í•˜ëŠ” í•­ëª©ì€ ì—†ìœ¼ë¯€ë¡œ ì œì™¸í•˜ê² ìŠµë‹ˆë‹¤.
# ë³´ê¸° ì‰½ê²Œ ë°ì´í„° í”„ë ˆì„ì— ë‹´ìŠµë‹ˆë‹¤.

df_names = pd.DataFrame({'ì¢…ëª©ëª…': ['í•˜ì´íŠ¸ì§„ë¡œ', 'ë³´í•´ì–‘ì¡°', 'ë¡¯ë°ì¹ ì„±', 'ë§ˆë‹ˆì»¤', 'ì´ì§€í™€ë”©ìŠ¤', 'í•˜ë¦¼']})
df_names
````

````python
# ê°ê°ì˜ ë°ì´í„° í”„ë ˆì„ì—ì„œ í•„ìš”í•œ ì‘ì—…ì„ í•˜ëŠ” í•¨ìˆ˜ë¥¼ ì •ì˜í•©ë‹ˆë‹¤.
# í•„ìš”í•œ ì»¬ëŸ¼ 3ê°œ ì„ íƒ > ë§ˆì§€ë§‰ ë¶„ê¸°ì— í•´ë‹¹í•˜ëŠ” í–‰ ì„ íƒ

def choose_row(df):
  # ì›ë³¸ ë°ì´í„° ë³´í˜¸ë¥¼ ìœ„í•´ copy
  df_tmp = df.copy()

  # í•„ìš”í•œ ì»¬ëŸ¼ 3ê°œ ì„ íƒ
  df_tmp = df_tmp[["ë§¤ì¶œì•¡", "ìì‚°ì´ê³„", "EPS(ì›)"]]
  
  # ë§ˆì§€ë§‰ ë¶„ê¸°ì— í•´ë‹¹í•˜ëŠ” í–‰ = ë§¨ ë§ˆì§€ë§‰ í–‰ ì„ íƒ
  df_tmp = pd.DataFrame(df_tmp.iloc[-1]).transpose()
  
  # ì™„ì„±ëœ ë°ì´í„° í”„ë ˆì„ì„ ë‚´ê°€ ì§€ì •í•œ ë³€ìˆ˜ì— ì €ì¥í•  ìˆ˜ ìˆë„ë¡ ë°˜í™˜í•©ë‹ˆë‹¤. 
  return df_tmp
````

````python
# 6ê°œì˜ ë°ì´í„° í”„ë ˆì„ì„ ê°€ì§€ê³  í•„ìš”í•œ ì‘ì—…ì„ í•©ë‹ˆë‹¤.
df1_f = choose_row(df1)
df2_f = choose_row(df2)
df3_f = choose_row(df3)
df4_f = choose_row(df4)
df5_f = choose_row(df5)
df6_f = choose_row(df6)
````

````python
# 6ê°œ & df_names ë°ì´í„° í”„ë ˆì„ì„ ëª¨ë‘ í•©ì¹©ë‹ˆë‹¤.
df = pd.concat([df1_f, df2_f, df3_f, df4_f, df5_f, df6_f])
df
````

````python
# ì¸ë±ìŠ¤ë¥¼ ë¬¸ì œì— ë§ì¶”ì–´ ë¬¸ìì—´ ì²˜ë¦¬ë¥¼ í•©ë‹ˆë‹¤.
index_list = []
for i in df.index:
  index_list.append(i.split("\n")[0])

index_list
````

````python
# ì¸ë±ìŠ¤ë¥¼ ì¬ì •ë ¬í•©ë‹ˆë‹¤. 
df.reset_index(drop=True, inplace=True)
df
````

````python
# dfì™€ df_namesì„ í•©ì¹©ë‹ˆë‹¤.
df = pd.concat([df, df_names], axis=1)
df
````

## 3. Merge & Join +

pandas mergeì˜ optionsì¤‘ how ì—ëŠ” ì—¬ëŸ¬ê°€ì§€ ì˜µì…˜ë“¤ì´ ìˆìŠµë‹ˆë‹¤.

ê³µì‹ ë ˆí¼ëŸ°ìŠ¤ë¥¼ ê²€ìƒ‰, ì°¸ì¡°í•˜ì—¬

ì•„ë˜ ì½”ë“œë¥¼ ë°”íƒ•ìœ¼ë¡œ 

ë‹¤ìŒê³¼ ê°™ì€ ê²°ê³¼ë¬¼ì„ ë§Œë“œëŠ” ì½”ë“œë¥¼ ì‘ì„±í•˜ì„¸ìš”

ë³€ìˆ˜ì˜ ì´ë¦„ì€ `df2` ì…ë‹ˆë‹¤.

|key1|key2|A|B|C|D|
|:-:|:-:|:-:|:-:|:-:|:-:|
|K0|K0|A0|B0|C0|D0|
|K0|K1|A1|B1|NaN|NaN|
|K1|K0|A2|B2|C1|D1|
|K1|K0|A2|B2|C2|D2|
|K2|K1|A3|B3|NaN|NaN|





````python
# import pandas as pd

left = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                     'key2': ['K0', 'K1', 'K0', 'K1'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})

right = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                      'key2': ['K0', 'K0', 'K0', 'K0'],
                      'C': ['C0', 'C1', 'C2', 'C3'],
                      'D': ['D0', 'D1', 'D2', 'D3']})
````

````python
# í™•ì¸ 1
left
````

````python
# í™•ì¸ 2
right
````

````python
# merge ì‹¤í–‰ > leftë¥¼ ì‚¬ìš©í•©ë‹ˆë‹¤.
# ìì„¸í•œ ê²ƒì€ python merge, python merge ì¢…ë¥˜ ë“±ì„ í‚¤ì›Œë“œë¡œ ê²€ìƒ‰í•´ë´…ë‹ˆë‹¤.
df2 = left.merge(right, how='left', on=['key1', 'key2'])
df2
````

## Grading

ì´ì „ ê³¼ì œë¥¼ ì™„ë£Œí–ˆë‹¤ë©´, ì•„ë˜ì˜ ì½”ë“œë¥¼ ì‹¤í–‰í•˜ì„¸ìš”. 

````python
import numpy as np

def grade(df, df2, df_tidy, res):
  score = 0
  score += (df['ë§¤ì¶œì•¡'][2] == 5980)
  score += (df['ìì‚°ì´ê³„'][3] == 1897)   
  score += (df2.equals(pd.DataFrame({
    'key1' : ['K0', 'K0', 'K1', 'K1', 'K2'],
    'key2' : ['K0', 'K1', 'K0', 'K0', 'K1'],
    'A' : ['A0', 'A1', 'A2', 'A2', 'A3'],
    'B' : ['B0', 'B1', 'B2', 'B2', 'B3'],
    'C' : ['C0', np.NaN, 'C1', 'C2', np.NaN],
    'D' : ['D0', np.NaN, 'D1', 'D2', np.NaN],
  }))) 
  score += (res == 6)
  score += (round(df_tidy['value'].mean(), 3) == 7090.444)
  return(score)
````

````python
grade(df, df2, df_tidy, res)
````
