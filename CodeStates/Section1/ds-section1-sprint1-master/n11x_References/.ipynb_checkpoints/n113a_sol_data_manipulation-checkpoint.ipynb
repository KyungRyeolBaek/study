````python
# 판다스
import pandas as pd
````

````python
# url 준비
urlhead = 'https://ds-lecture-data.s3.ap-northeast-2.amazonaws.com/stocks/'
url_list = ['000080.csv','000890.csv', '005300.csv', '027740.csv', '035810.csv', '136480.csv']
url_final = []

for i in url_list:
  url = urlhead + i
  url_final.append(url)
````

````python
# url으로 데이터를 불러와서 데이터 프레임에 담은 후, 원하는 모양대로 가공하는 함수를 정의합니다.
def dataframe_load(url):

  # url 불러오기
  df_tmp = pd.read_csv(url)
  
  # transpose
  df_tmp = df_tmp.T
  
  # 첫 번째 행을 인덱스로 설정
  new_header = df_tmp.iloc[0]
  df_tmp.columns = new_header

  # 첫 번째 행을 제외한 나머지 행들만 선택
  df_tmp = df_tmp[1:]

  # 완성된 데이터 프레임을 내가 지정한 변수에 저장할 수 있도록 반환합니다. 
  return df_tmp
````

````python
# 각각의 csv파일을 데이터 프레임으로 만들어 변수에 저장합니다.
df1 = dataframe_load(url_final[0])
df2 = dataframe_load(url_final[1])
df3 = dataframe_load(url_final[2])
df4 = dataframe_load(url_final[3])
df5 = dataframe_load(url_final[4])
df6 = dataframe_load(url_final[5])
````

````python
df4
````

````python
# 학습 노트에서 쓰던 데이터를 계속 쓰는 것 이므로 종목명을 그대로 가져옵니다.
# csv 번호를 살펴보니 136490번에 해당하는 항목은 없으므로 제외하겠습니다.
# 보기 쉽게 데이터 프레임에 담습니다.

df_names = pd.DataFrame({'종목명': ['하이트진로', '보해양조', '롯데칠성', '마니커', '이지홀딩스', '하림']})
df_names
````

````python
# 각각의 데이터 프레임에서 필요한 작업을 하는 함수를 정의합니다.
# 필요한 컬럼 3개 선택 > 마지막 분기에 해당하는 행 선택

def choose_row(df):
  # 원본 데이터 보호를 위해 copy
  df_tmp = df.copy()

  # 필요한 컬럼 3개 선택
  df_tmp = df_tmp[["매출액", "자산총계", "EPS(원)"]]
  
  # 마지막 분기에 해당하는 행 = 맨 마지막 행 선택
  df_tmp = pd.DataFrame(df_tmp.iloc[-1]).transpose()
  
  # 완성된 데이터 프레임을 내가 지정한 변수에 저장할 수 있도록 반환합니다. 
  return df_tmp
````

````python
# 6개의 데이터 프레임을 가지고 필요한 작업을 합니다.
df1_f = choose_row(df1)
df2_f = choose_row(df2)
df3_f = choose_row(df3)
df4_f = choose_row(df4)
df5_f = choose_row(df5)
df6_f = choose_row(df6)
````

````python
# 6개 & df_names 데이터 프레임을 모두 합칩니다.
df = pd.concat([df1_f, df2_f, df3_f, df4_f, df5_f, df6_f])
df
````

````python
# 인덱스를 문제에 맞추어 문자열 처리를 합니다.
index_list = []
for i in df.index:
  index_list.append(i.split("\n")[0])

index_list
````

````python
# 인덱스를 재정렬합니다. 
df.reset_index(drop=True, inplace=True)
df
````

````python
# df와 df_names을 합칩니다.
df = pd.concat([df, df_names], axis=1)
df
````

````python
# 아까 저장해 둔 인덱스 리스트르 바탕으로 df의 인덱스 재설정
df.index = index_list
````

다음과 같은 결과물을 만드는 코드를 작성하세요

변수의 이름은 `df2` 입니다.

|key1|key2|A|B|C|D|
|:-:|:-:|:-:|:-:|:-:|:-:|
|K0|K0|A0|B0|C0|D0|
|K0|K1|A1|B1|NaN|NaN|
|K1|K0|A2|B2|C1|D1|
|K1|K0|A2|B2|C2|D2|
|K2|K1|A3|B3|NaN|NaN|





````python
# import pandas as pd

left = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                     'key2': ['K0', 'K1', 'K0', 'K1'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})

right = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                      'key2': ['K0', 'K0', 'K0', 'K0'],
                      'C': ['C0', 'C1', 'C2', 'C3'],
                      'D': ['D0', 'D1', 'D2', 'D3']})
````

````python
# 확인 1
left
````

````python
# 확인 2
right
````

````python
# merge 실행 > left를 사용합니다.
# 자세한 것은 python merge, python merge 종류 등을 키워드로 검색해봅니다.
df2 = left.merge(right, how='left', on=['key1', 'key2'])
df2
````

## Grading

이전 과제를 완료했다면, 아래의 코드를 실행하세요. 

````python
import numpy as np

def grade(df, df2, df_tidy, res):
  score = 0
  score += (df['매출액'][2] == 5980)
  score += (df['자산총계'][3] == 1897)   
  score += (df2.equals(pd.DataFrame({
    'key1' : ['K0', 'K0', 'K1', 'K1', 'K2'],
    'key2' : ['K0', 'K1', 'K0', 'K0', 'K1'],
    'A' : ['A0', 'A1', 'A2', 'A2', 'A3'],
    'B' : ['B0', 'B1', 'B2', 'B2', 'B3'],
    'C' : ['C0', np.NaN, 'C1', 'C2', np.NaN],
    'D' : ['D0', np.NaN, 'D1', 'D2', np.NaN],
  }))) 
  score += (res == 6)
  score += (round(df_tidy['value'].mean(), 3) == 7090.444)
  return(score)
````

````python
grade(df, df2, df_tidy, res)
````
