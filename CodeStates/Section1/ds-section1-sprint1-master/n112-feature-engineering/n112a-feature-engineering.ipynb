### Feature Engineering

Relative Perfomance ë¼ëŠ” ìƒˆë¡œìš´ featureë¥¼ ê³„ì‚°í•˜ì„¸ìš”.

ì´ëŠ” ìµœê·¼ 1ë…„ì¹˜ ë§¤ì¶œì•¡ì˜ í‰ê· ê°’ì„ ê¸°ì¤€ìœ¼ë¡œ 

- 10% ì´ìƒ -> `S`
- 5% ì´ìƒ -> `A`
- -5 ~ 5% -> `B`
- -5%ì´í•˜ -> `C`
- 10%ì´í•˜ -> `D`

ë¼ëŠ” ê°’ì„ ê°–ëŠ” featureì…ë‹ˆë‹¤.

20ë…„ë„ 2ë¶„ê¸°ì— í•´ë‹¹í•˜ëŠ” ê²°ê³¼ê°’ì€ `A`ê°€ ë‚˜ì™€ì•¼í•©ë‹ˆë‹¤.

ê°ê°ì— í•´ë‹¹í•˜ëŠ” ë“±ê¸‰ì´ ë‚˜ì˜¤ê¸° ìœ„í•´ì„œ í•„ìš”í•œ ë§¤ì¶œì•¡ì„ ì¶”ê°€ë¡œ ì„œìˆ í•˜ì„¸ìš”.

```
import numpy as np

RP = ( df['ë§¤ì¶œì•¡'] / df['ë§¤ì¶œì•¡'].mean() - 1 ) * 100

df['Relative Perfomance'] = ( df['ë§¤ì¶œì•¡'] / df['ë§¤ì¶œì•¡'].mean() - 1 ) * 100

df['Relative Perfomance'][RP<-10] = 'D' 
df['Relative Perfomance'][(RP>-10) & (RP<-5)]= 'C'
df['Relative Perfomance'][(RP>-5) & (RP<5)]= 'B'
df['Relative Perfomance'][(RP>5) & (RP<10)]= 'A'
df['Relative Perfomance'][RP>10]= 'S' 
df
```
