## 2. Blood Test

íŠ¹ì • ì§ˆë³‘ì„ ê°€ì§€ê³  ìˆëŠ” ê²½ìš° **99%** ì˜ í™•ë¥ ë¡œ íƒì§€ í•  ìˆ˜ ìˆëŠ” ì‹¤í—˜ë°©ë²•ì´ ìˆìŠµë‹ˆë‹¤.

ê·¸ëŸ¬ë‚˜ ë™ì‹œì— ì´ ë°©ë²•ì€, **1%** ì˜ í™•ë¥ ë¡œ ì§ˆë³‘ì´ ì—†ì§€ë§Œ ì§ˆë³‘ì´ ìˆë‹¤ê³  ì§„ë‹¨ í•˜ëŠ” ê²½ìš°ë„ ìˆëŠ”ë°ìš”.

ì‹¤ì œë¡œ ëª¨ë“  ì¸êµ¬ì¤‘ **0.5%** ë§Œì´ ì´ ì§ˆë³‘ì„ ê°€ì§€ê³  ìˆë‹¤ê³  í•˜ê² ìŠµë‹ˆë‹¤.

íŠ¹ì • ì‚¬ëŒì´ ì´ ë°©ë²•ì„ í†µí•´ ì§ˆë³‘ì´ ìˆë‹¤ê³  ì§„ë‹¨ ë°›ì•˜ì„ë•Œ, ì‹¤ì œë¡œ ì§ˆë³‘ì„ ê°€ì§€ê³  ìˆì„ í™•ë¥ ì„ êµ¬í•˜ëŠ” í•¨ìˆ˜ë¥¼ ì‘ì„±í•˜ì„¸ìš”.

```python
def disease(prior, tpr, fpr):
  # prior = 0.005 # = ì‚¬ì „ í™•ë¥  = ì§ˆë³‘ì´ ìˆì„ í™•ë¥ 
  # tpr = 0.99  # = ì‹¤ì œ ì°¸ì¸ê²ƒì„ ì°¸ì´ë¼ê³  íŒë‹¨í•˜ëŠ” í™•ë¥  = ë¯¼ê°ë„
  # fpr = 0.01  # = ì‹¤ì œ ê±°ì§“ì¸ê²ƒì„ ê±°ì§“ì´ë¼ íŒë‹¨í•˜ëŠ” í™•ë¥  = False ë¥¼ Trueë¡œ ì˜ëª» íŒë‹¨ í•˜ëŠ” í™•ë¥  = 1-íŠ¹ì´ë„
  # ì§ˆë³‘ì´ ìˆëŠ”ë° ì§ˆë³‘ì´ ìˆë‹¤ê³  íŒì • ë°›ì„ í™•ë¥  = 0.005 * 0.99 = prior * tpr
  # ì§ˆë³‘ì´ ì—†ëŠ”ë° ì§ˆë³‘ì´ ìˆë‹¤ê³  íŒì • ë°›ì„ í™•ë¥  = 0.995 * 0.01 = (1-prior) * fpr
  # P(A|B) = P(B|A)*P(A) / P(B) = P(B|A)*P(A) / P(B|A)*P(A) + P(B|A(c))*(A(c))
  # ì§ˆë³‘ì´ ìˆë‹¤ê³  ì§„ë‹¨ ë°›ì•˜ì„ ë•Œ, ì‹¤ì œë¡œ ì§ˆë³‘ì„ ê°€ì§€ê³  ìˆì„ í™•ë¥  = ì§ˆë³‘ì´ ìˆëŠ”ë° ì§ˆë³‘ì´ ìˆë‹¤ê³  íŒì • ë°›ì„ í™•ë¥  / ì§ˆë³‘ì´ ìˆëŠ”ë° ì§ˆë³‘ì´ ìˆë‹¤ê³  íŒì •ë°›ì„ í™•ë¥  + ì§ˆë³‘ì´ ì—†ëŠ”ë° ì§ˆë³‘ì´ ìˆë‹¤ê³  íŒì • ë°›ì„ í™•ë¥ 
  return prior * tpr / (prior * tpr + (1-prior) * fpr )
  # ì½”ë“œë¥¼ ì‘ì„±í•˜ì„¸ìš”

disease(0.005, 0.99, 0.01)
```

```python
# ì´ˆê¸° ì¡°ì‚¬
# ìš©ì˜ìê°€ ë²”ì£„ë¥¼ ì €ì§ˆë €ì„ í™•ë¥  0.6
# ë²”ì¸ì´ ì™¼ì†ì¡ì´ì´ë‹¤
# ì™¼ì†ì¡ì´ì˜ ì¸êµ¬ìˆ˜ 0.2
# ìš©ì˜ìê°€ ë²”ì£„ë¥¼ ì €ì§ˆë €ë‹¤ / ìš©ì˜ìëŠ” ì™¼ì†ì¡ì´ë‹¤.
# ìš©ì˜ìê°€ ë²”ì¸ì¼ í™•ë¥  = ìš©ì˜ìê°€ ë²”ì£„ë¥¼ ì €ì§ˆë €ì„ í™•ë¥  * ìš©ì˜ìê°€ ì™¼ì†ì¼ í™•ë¥  = 0.6 * 1 = 0.6
# ìš©ì˜ìê°€ ë²”ì¸ì´ ì•„ë‹ˆê³  ë‹¤ë¥¸ ì™¼ì†ì¡ì´ê°€ ë²”ì¸ì¼ í™•ë¥  = 0.4 * 0.2
# P(ìš©ì˜ì) = 0.6 = ìš©ì˜ìì¼ í™•ë¥ 
# P(ì™¼ì†|ìš©ì˜ì(c)) = 0.2 -> ë¬¸ì œì—ì„œ ì£¼ì–´ì§ = ìš©ì˜ì ì™¸ì˜ ì‚¬ëŒì´ ì™¼ì†ì¼ í™•ë¥ 
# P(ì™¼ì†|ìš©ì˜ì) = 1 -> ë¬¸ì œì—ì„œ ì£¼ì–´ì§ = ìš©ì˜ìê°€ ì™¼ì†ì¼ í™•ë¥ 
# P(ìš©ì˜ì(c)) = 0.4 -> 1 - P(ìš©ì˜ì) = ìš©ì˜ìê°€ ì•„ë‹ í™•ë¥ 
# P(ìš©ì˜ì|ì™¼ì†) = P(ì™¼ì†|ìš©ì˜ì)P(ìš©ì˜ì) / P(ì™¼ì†) = P(ì™¼ì†|ìš©ì˜ì)P(ìš©ì˜ì) / P(ì™¼ì†|ìš©ì˜ì)P(ìš©ì˜ì) + P(ì™¼ì†|ìš©ì˜ì(c))P(ìš©ì˜ì(c))
# 1*0.6 / 1*0.6 + 0.2*0.4
pre = 1*0.6 / ((1*0.6) + (0.2*0.4))
print('ì´ˆê¸° ì¡°ì‚¬ : {}, í™•ë¥  : {}%'.format(pre, pre*100))
# ì´ í›„ ì¡°ì‚¬
# ì™¼ì†ì¼ í™•ë¥  90%
# P(ì™¼ì†|ìš©ì˜ì) = 0.9 -> ë¡œ ë°”ë€”ê²½ìš°! ë‚˜ë¨¸ì§€ëŠ” ë³€í•¨ì´ ì—†ìŒ. ê·¸ëŒ€ë¡œ ì‹ì— ë„£ì–´ë³´ë©´
# 0.9*0.6 / (0.9*0.6 + 0.2*0.4)
after = 0.9*0.6 / (0.9*0.6 + 0.2*0.4)
print('ì´í›„ ì¡°ì‚¬ : {}, í™•ë¥  : {}%'.format(after, after*100))

# ì´ ê³³ì— ë‹µì•ˆì„ ì‘ì„±í•˜ì‹œê¸° ë°”ëë‹ˆë‹¤
```
