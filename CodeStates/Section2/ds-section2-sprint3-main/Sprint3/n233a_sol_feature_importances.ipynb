### 2) Xgboostë¥¼ ì‚¬ìš©í•´ í•™ìŠµí•˜ê³  ê¸°ì¡´ ëª¨ë¸ê³¼ ë¹„êµí•´ ë³´ì„¸ìš”.
- ë³¸ì¸ í¬íŠ¸í´ë¦¬ì˜¤ ë°ì´í„°ë¥¼ Xgboostë¡œ í•™ìŠµí•˜ì—¬ ê¸°ì¡´ì— ë§Œë“¤ì—ˆë˜ ëª¨ë¸ê³¼ ë¹„êµí•©ë‹ˆë‹¤.
- **breast cancer ë°ì´í„°ë¥¼ Xgboostë¡œ í•™ìŠµí•˜ê³  f1 scoreë¥¼ ì œì¶œí•˜ì„¸ìš”.(test ë°ì´í„°ì—ì„œ 0.95ì´ìƒ)**

```python
### (ë³¸ì¸ í¬íŠ¸í´ë¦¬ì˜¤ ê³¼ì œ) ì´ê³³ì—ì„œ ê³¼ì œë¥¼ ì§„í–‰í•´ ì£¼ì„¸ìš” ### 
```

```python
### (ì œì¶œí¼ ê³¼ì œ) ì´ê³³ì—ì„œ ê³¼ì œë¥¼ ì§„í–‰í•´ ì£¼ì„¸ìš” ###

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_breast_cancer 

cancer_dataset = load_breast_cancer()

features = cancer_dataset.data
target = cancer_dataset.target
df = pd.DataFrame(data=features, columns=cancer_dataset.feature_names)
df['target'] = target

X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=2)
print(X_train.shape, X_test.shape)
```

```python
from sklearn.pipeline import make_pipeline
from category_encoders import OrdinalEncoder
from xgboost import XGBClassifier

pipe = make_pipeline(
    OrdinalEncoder(),
    XGBClassifier(n_estimators=200
                  , random_state=2
                  , n_jobs=-1
                  , max_depth=7
                  , learning_rate=0.2
                 )
)

pipe.fit(X_train, y_train);
```

```python
from sklearn.metrics import accuracy_score, classification_report, f1_score
y_pred = pipe.predict(X_test)
print('\n <test ì •í™•ë„> \n', accuracy_score(y_test, y_pred))

print("\n <classification_report> \n ", classification_report(y_pred, y_test))

print("\n <f1 score> \n ", f1_score(y_pred, y_test))

# f1_score : 0.9635036496350365
```

```python
### (ë³¸ì¸ í¬íŠ¸í´ë¦¬ì˜¤ ê³¼ì œ) ì´ê³³ì—ì„œ ê³¼ì œë¥¼ ì§„í–‰í•´ ì£¼ì„¸ìš” ### 
```

```python
import eli5
from eli5.sklearn import PermutationImportance

pipe.named_steps
```

```python
permuter = PermutationImportance(
    pipe.named_steps['xgbclassifier'],
    scoring='accuracy', 
    n_iter=5,
    random_state=2
)
```

```python
X_test_transformed = pipe.named_steps['ordinalencoder'].transform(X_test)

permuter.fit(X_test_transformed, y_test);
```

```python
feature_names = cancer_dataset.feature_names.tolist()
pd.Series(permuter.feature_importances_, feature_names).sort_values()
```

```python
eli5.show_weights(
    permuter, 
    top=None, # top n ì§€ì • ê°€ëŠ¥, None ì¼ ê²½ìš° ëª¨ë“  íŠ¹ì„± 
    feature_names=feature_names
)


# ê°€ì¥ ë†’ì€ ê°’ì„ ê°€ì§€ëŠ” íŠ¹ì„±ì˜ ì´ë¦„ : mean smoothness
```
