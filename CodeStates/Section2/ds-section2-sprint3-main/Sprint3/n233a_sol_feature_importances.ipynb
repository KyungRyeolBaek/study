### 2) Xgboost를 사용해 학습하고 기존 모델과 비교해 보세요.
- 본인 포트폴리오 데이터를 Xgboost로 학습하여 기존에 만들었던 모델과 비교합니다.
- **breast cancer 데이터를 Xgboost로 학습하고 f1 score를 제출하세요.(test 데이터에서 0.95이상)**

```python
### (본인 포트폴리오 과제) 이곳에서 과제를 진행해 주세요 ### 
```

```python
### (제출폼 과제) 이곳에서 과제를 진행해 주세요 ###

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_breast_cancer 

cancer_dataset = load_breast_cancer()

features = cancer_dataset.data
target = cancer_dataset.target
df = pd.DataFrame(data=features, columns=cancer_dataset.feature_names)
df['target'] = target

X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=2)
print(X_train.shape, X_test.shape)
```

```python
from sklearn.pipeline import make_pipeline
from category_encoders import OrdinalEncoder
from xgboost import XGBClassifier

pipe = make_pipeline(
    OrdinalEncoder(),
    XGBClassifier(n_estimators=200
                  , random_state=2
                  , n_jobs=-1
                  , max_depth=7
                  , learning_rate=0.2
                 )
)

pipe.fit(X_train, y_train);
```

```python
from sklearn.metrics import accuracy_score, classification_report, f1_score
y_pred = pipe.predict(X_test)
print('\n <test 정확도> \n', accuracy_score(y_test, y_pred))

print("\n <classification_report> \n ", classification_report(y_pred, y_test))

print("\n <f1 score> \n ", f1_score(y_pred, y_test))

# f1_score : 0.9635036496350365
```

### 3) Permutation importances를 계산하고 다른 중요도 결과와 비교해 보세요
- 본인 포트폴리오 데이터를 사용해 만든 모델의 Permutation importances를 계산하세요.
- **breast cancer 데이터를 사용해 만든 모델의 Permutation importances를 계산하고 가장 높은 값을 가지는 특성의 이름을 과제 제출폼에 제출하세요.**
    - eli5 PermutationImportance 사용
    - scoring = 'accuracy'
    - n_iter = 5
    - random_state = 2

```python
### (본인 포트폴리오 과제) 이곳에서 과제를 진행해 주세요 ### 
```

```python
import eli5
from eli5.sklearn import PermutationImportance

pipe.named_steps
```

```python
permuter = PermutationImportance(
    pipe.named_steps['xgbclassifier'],
    scoring='accuracy', 
    n_iter=5,
    random_state=2
)
```

```python
X_test_transformed = pipe.named_steps['ordinalencoder'].transform(X_test)

permuter.fit(X_test_transformed, y_test);
```

```python
feature_names = cancer_dataset.feature_names.tolist()
pd.Series(permuter.feature_importances_, feature_names).sort_values()
```

```python
eli5.show_weights(
    permuter, 
    top=None, # top n 지정 가능, None 일 경우 모든 특성 
    feature_names=feature_names
)


# 가장 높은 값을 가지는 특성의 이름 : mean smoothness
```
